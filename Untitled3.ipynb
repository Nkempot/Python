```python
# a) Strip whitespace from 'Country/Territory'
df['Country/Territory'] = df['Country/Territory'].str.strip()

# b) Convert 'IMF_Year' to datetime
df['IMF_Year'] = pd.to_datetime(df['IMF_Year'], format='%Y', errors='coerce')

# c) Convert 'WorldBank_Estimate' to integer -- if it was not already an int
df['WorldBank_Estimate'] = pd.to_numeric(df['WorldBank_Estimate'], errors='coerce').fillna(0).astype(int)
```
